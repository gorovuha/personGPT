```python
embed = pd.DataFrame()
#.strip("]'[").split("', '")
for i in range(2, 7):
    row = []
    for j in range(4):
        emb = []
        for k in range(10):
            temp = model.encode(data[data.columns[i]][j].strip("]'[").split("', '")[k].lower())
            emb.append(temp)
        row.append(emb)
    embed[data.columns[i]] = row
```

```python
import numpy as np
from numpy import dot
from numpy.linalg import norm
from scipy import spatial

person_id = []
person_name = []
assoc = []
term1 = []
term2 = []
cos = []
pers_id2 = []
pers_name2 = []

#print(data['университет'][0])
#print((data['университет'][0].strip("]'[").split("', '"))[0])

n = ['enrollee', 'graduate', 'parent','post-doc']

w = ['center', 'university', 'scientific/science', 'educational/education', 'program']

for i in range(5):
    for j in range(4):
        for k in range(10):
            for l in range(4):
                for m in range(10):
                    _ = 1 - spatial.distance.cosine(embed[w[i]][j][k], embed[w[i]][l][m])
                    if k!=m or l!=j:
                        person_id.append(j)
                        person_name.append(n[j])
                        assoc.append(w[i])
                        term1.append(data[w[i]][j].strip("]'[").split("', '")[k])
                        term2.append(data[w[i]][l].strip("]'[").split("', '")[m])
                        cos.append(_)
                        pers_id2.append(l)
                        pers_name2.append(n[l])
```

```python
pers_age1 = []
pers_gen1 = []
pers_age2 = []
pers_gen2 = []
edu = []
edu2 = []

for i in range(len(person_id)):
    if person_id[i] == 0:
        pers_age1.append(18)
        pers_gen1.append('female')
        edu.append(1)
    if person_id[i] == 1:
        pers_age1.append(25)
        pers_gen1.append('male')
        edu.append(2)
    if person_id[i] == 2:
        pers_age1.append(40)
        pers_gen1.append('female')
        edu.append(2)
    if person_id[i] == 3:
        pers_age1.append(34)
        pers_gen1.append('male')
        edu.append(3)
    if pers_id2[i] == 0:
        pers_age2.append(18)
        pers_gen2.append('female')
        edu2.append(1)
    if pers_id2[i] == 1:
        pers_age2.append(25)
        pers_gen2.append('male')
        edu2.append(2)
    if pers_id2[i] == 2:
        pers_age2.append(40)
        pers_gen2.append('female')
        edu2.append(2)
    if pers_id2[i] == 3:
        pers_age2.append(34)
        pers_gen2.append('male')
        edu2.append(3)
```

```python
df = pd.DataFrame()
df['Person_ID'] = person_id
df['Person_name'] = person_name
df['Person_age'] = pers_age1
df['Person_gender'] = pers_gen1
df['Person_ducation_level_ID'] = edu
df['Assoc_stimulus'] = assoc
df['Term1'] = term1
df['Term2'] = term2
df['cos'] = cos
df['Person_term_ID'] = pers_id2
df['Person_term_name'] = pers_name2
df['Person_term_age'] = pers_age2
df['Person_term_gender'] = pers_gen2
df['Person_term_ducation_level_ID'] = edu2
```
